```python
from qiskit import QuantumCircuit
from qiskit import execute, Aer
from qiskit.circuit import Parameter
from qiskit.visualization import plot_histogram
```

### Quantum Gates Example

```python
q_circ1=QuantumCircuit(1)
q_circ1.h(0)
q_circ1.draw(output='mpl')
```

```python
q_circ2=QuantumCircuit(2,2)
q_circ2.cx(0,1)
q_circ2.draw(output='mpl')
```

### Hello, Quantum

```python

Q_demo=QuantumCircuit(2,2)


Q_demo.x(0)
Q_demo.barrier()
Q_demo.cx(0,1)
Q_demo.barrier()
Q_demo.measure([0,1],[0,1])

Q_demo.draw(output='mpl')
```

```python
# Use Aer's qasm_simulator
simulator=Aer.get_backend('qasm_simulator')

# Execute and grab the results of the circuit on the qasm simulator with 5000 repeats of circuit, default is 1024.
result=execute(Q_demo,simulator,shots=5000).result()

# Print the counts and plot the histogram of circuit.
print(result.get_counts())
plot_histogram(result.get_counts())
```

```python
Q_bell=QuantumCircuit(2,2)


Q_bell.h(0)
Q_bell.barrier()
Q_bell.cx(0,1)
Q_bell.barrier()
Q_bell.measure([0,1],[0,1])

Q_bell.draw(output='mpl')
```
